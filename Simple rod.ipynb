```python
import numpy as np
from math import log
from random import *

n0=100;#number of particles per cycle
run=10;#Number of runs
#starting location of particles
loc0=0;
#macroscopic cross sections [cm-1]
sigc=0.1643
sigs=0.4011
sigf=0.2030
sigt=sigc+sigs+sigf
siga=sigc+sigf

#nubar
nubar=2.436700e+0; #U-235 Thermal
#nubar=2.492088e+0; #U-238 Thermal
#nubar=2.878900e+0; #Pu-239 Thermal
#length of the rod
length=10
binwidth=.1
bins=length/binwidth
bin0=loc0/binwidth
```

```python
n=n0;#number of particles 
loc=loc0
nrun=0
x=np.zeros(np.int(bins) )#bins for flux calculations
fbank=np.zeros(np.int((bins)))
fbank[np.int(bin0)]=n0 #fission neutron bank
#temporary locations for created fission events
absorb=0
leak=0

while n>0:
    move=-log(random())/sigt;#how far the particle moves
    omega=2*random()-1;direction=omega/abs(omega) 
    move=move*omega
    if length<loc+move  or   loc+move<0:
        leak=leak+1
        n=n-1
        loc=loc0
    else:
        rn=random();# number to determine the reaction type
        if rn<sigc/sigt:#capture
            absorb =absorb+1
            n=n-1
            loc=loc0
        elif rn<(sigc+sigf)/sigt:#fission
            if rn<0.0317223:
                n=n-1
                loc=loc0
            elif rn<0.2034294:
                n=n-1
                fbank[np.int(round((move+loc)/binwidth,0))]=fbank[np.int(round((move+loc)/binwidth))]+1
                loc=loc0
            elif rn<0.5396285:
                n=n-1
                fbank[np.int(round((move+loc)/binwidth))]=fbank[np.int(round((move+loc)/binwidth))]+2
                loc=loc0
            elif rn<0.843598:
                n=n-1
                fbank[np.int(round((move+loc)/binwidth))]=fbank[np.int(round((move+loc)/binwidth))]+3
                loc=loc0
            elif rn<0.9705439:
                n=n-1
                fbank[np.int(round((move+loc)/binwidth))]=fbank[np.int(round((move+loc)/binwidth))]+4
                loc=loc0
            elif rn<0.9972232:
                n=n-1
                fbank[np.int(round((move+loc)/binwidth))]=fbank[np.int(round((move+loc)/binwidth))]+5
                loc=loc0
            elif rn<0.9998554:
                n=n-1
                fbank[np.int(round((move+loc)/binwidth))]=fbank[np.int(round((move+loc)/binwidth))]+6
                loc=loc0
            else:
                n=n-1
                fbank[np.int(round((move+loc)/binwidth))]=fbank[np.int(round((move+loc)/binwidth))]+7
                loc=loc0
        else:#scatter
            loc=loc+move
            
i=0
while sum(fbank)>n0:
    if i>=len(fbank-1):
        i=0
    rx=random()
    if rx>.5 and fbank[i]>0:
        fbank[i]=fbank[i]-1
        i=i+1
    else:
        i=i+1
i=0
while sum(fbank)<n0:
    if i>=len(fbank-1):
        i=0
    rx=random()
    if rx>.5:
        fbank[i]=fbank[i]+1
        i=i+1
    else:
        i=i+1        
print((fbank))           
            
        
```
